This notebook is meant to be an improvemnt over my existing method for estimating the galaxy centroid uncertainty.  Currently I am overly optimistic in my estimate of the centroid uncertainty.  Currently I treat any galaxies with spec-z within 2 times the cluster velocity dispersion as being confirmed clusters members and these remain fixed members during the bootstrap process.  Thus I am only randomly sampling from the remiander of the galaxies with photo-z in the range 0.43-0.63.  By doing this I think that I am not capturing the Poisson sampling noise inherent in the galaxy centroid estimate.

Note that this does assume that there is some underlying parent distribution of galaxies which we believe is what actually should coincide with the dark matter distribution, and that the observed galaxies are a fair Poisson sample of this underlying distribution.

## For spectroscopic galaxies

Given that the typical spectroscopic redshift uncertainty is ~0.00002, the spectroscopic redshifts can be approximated as delta functions.

```python
%%latex
$p(draw)_i = \frac{1}{\sigma_{\rm vdisp}\sqrt{2\pi}}e^{\frac{-(z_{\rm spec_{i}}-z_{\rm cluster})^2}{2\sigma_{\rm vdisp}^2}}$
```

## For photometric redshift galaxies

```python
%%latex
$p(draw)_i = \int_{0}^{\infty} p(z)_i \frac{1}{\sigma_{\rm vdisp}\sqrt{2\pi}}e^{\frac{-(z-z_{\rm cluster})^2}{2\sigma_{\rm vdisp}^2}}\mathrm{d}z$
```

```python
%%latex
$p(draw)_i = \int_{0}^{\infty} \frac{1}{\sigma_{\rm z-phot_{i}}\sqrt{2\pi}}e^{\frac{-(z-z_{\rm phot_{i}})^2}{2\sigma_{\rm z-phot_{i}}^2}} \frac{1}{\sigma_{\rm vdisp}\sqrt{2\pi}}e^{\frac{-(z-z_{\rm cluster})^2}{2\sigma_{\rm vdisp}^2}}\mathrm{d}z$
```

```python
# typical cluster redshift range
(1+0.53)*2000/3e5
```

Given that the typical cluster redshift range is the size of a p(z) photo-z bin there isn't much value in convolving the full p(z) with the cluster velocity dispersion.  Instead I could either take the p(z) value nearest the cluster redshift or I could treat the galaxy p(z) as a Gaussian distribution and sample from that.

#Functions

```python
def p_z_phot(z,z_phot,sigma_z_phot,z_cluster,sigma_z_cluster):
    return 1/(sigma_z_phot*numpy.sqrt(2*pi))*numpy.exp(-(z-z_phot)**2/(2*sigma_z_phot**2)) * 1/(sigma_z_cluster*numpy.sqrt(2*pi))*numpy.exp(-(z-z_cluster)**2/(2*sigma_z_cluster**2))

def weight_z_phot(z_phot,sigma_z_phot,z_cluster,sigma_z_cluster):
    """
    Usage: weight_z_phot(z,z_phot,sigma_z_phot,z_cluster,sigma_z_cluster)

    Returns the joint probability of overlapping galaxy and cluster redshift Gaussian PDF's.
    """
    prob = scipy.integrate.quad(lambda x: p_z_phot(x,z_phot,sigma_z_phot,z_cluster,sigma_z_cluster),0,inf)[0]
    return prob
```

```python
def p_z_spec(z,z_cluster,sigma_z_cluster):
    return 1/(sigma_z_cluster*numpy.sqrt(2*pi))*numpy.exp(-(z-z_cluster)**2/(2*sigma_z_cluster**2))

def weight_z_spec(z,z_cluster,sigma_z_cluster):
    """
    Usage: weight_z_phot(z,z_phot,sigma_z_phot,z_cluster,sigma_z_cluster)

    Returns the joint probability of overlapping galaxy and cluster redshift Gaussian PDF's.
    """
    prob = 1/(sigma_z_cluster*numpy.sqrt(2*pi))*numpy.exp(-(z-z_cluster)**2/(2*sigma_z_cluster**2))
    return prob
```

```python
def plotgaus(z_phot,sigma_z_phot,z_cluster,sigma_z_cluster):
    fig = figure()
    z = arange(0,1,0.001)
    phot = 1/(sigma_z_phot*numpy.sqrt(2*pi))*numpy.exp(-(z-z_phot)**2/(2*sigma_z_phot**2))
    clust = 1/(sigma_z_cluster*numpy.sqrt(2*pi))*numpy.exp(-(z-z_cluster)**2/(2*sigma_z_cluster**2))
    
    plot(z,phot,label='$p(z)_{galaxy}$')
    plot(z,clust,label='$p(z)_{cluster}$')
    plot(z,phot*clust,label='$p_{galaxy}$ x $p_{cluster}$')
    legend(loc=0)
    xlabel('z')
```

Obviously this works to eliminate the majority of the outliers but I don't really like this solution.  For one we know that there is a physical reason for there being a distribution of spectroscopic redshifts around the cluster center. Should we really favor those galaxies with spec-z at the peak of the distribution vs. those at the 1-sigma mark?  I don't think so.  Thus I think that I should assign a weight of 1 for all spec-z galaxies within 3-sigma of the cluster redshift.  Then I think that I should assign a weight to each photo-z galaxy proportional to the probability p(draw)_i normalized by the probability for the case where a photo-z galaxy is at the cluster redshift (since this is the case that will have the highest weight).

# Investigating New Weighting Scheme

```python
def weighting_mod(cat,z_cluster,sigma_z_cluster):
    '''
    calculate the redshift membership weights associated with each galaxy
    '''
    N_cat = numpy.shape(cat)[0]
    sigma_z_phot = s*(1+cat[:,key['z_mid']])
    # calculate the weight of a photo-z galaxy at the cluster redshift
    sigma_z_phot_clsuter = s*(1+z_cluster)
    weight_z_phot_norm = weight_z_phot(z_cluster,sigma_z_phot_cluster,z_cluster,sigma_z_cluster)
    # determine the weights for all galaxies
    weight = numpy.zeros(N_cat)
    for i in range(N_cat):
        if cat[i,key['z_spec']] == -99:
            weight[i] = weight_z_phot(cat[i,key['z_mid']],sigma_z_phot[i],z_cluster,sigma_z_cluster)/weight_z_phot_norm
        else:
            weight[i] = 1
    # set any weight values that might be negative = 0, these seem to be related to integration uncertainty
    mask_posweight = weight >= 0
    weight *= mask_posweight
    return weight
```

# Photo-z Penalizatoin Factor

This penalization factor is based in the fact that from our "magnitude limited" survey (R<23.5) of the Musket Ball we targeted 355 galaxies between 0.43= 3) and 210 of these spectra were between 0.525<specz<0.54 (with quality>=3), see for example the figure below. This translates to a sample purity of 63%.  Thus I will multiply all of the photo-z galaxy weights by a penalization factor of 0.63.

```python
def weighting_pen(cat,z_cluster,sigma_z_cluster,photo_z_penalty=0.63):
    '''
    calculate the redshift membership weights associated with each galaxy
    '''
    N_cat = numpy.shape(cat)[0]
    # calculate the approximate photo-z uncertainty for each galaxy
    sigma_z_phot = s*(1+cat[:,key['z_mid']])
    # Determine where the photo-z weight is maximized
    z_test = numpy.arange(z_cluster-0.2,z_cluster+0.2,0.001)
    sigma_z_phot_test = s*(1+z_test)
    N_test = numpy.size(z_test)
    weight_test = numpy.zeros(N_test)
    for i in numpy.arange(N_test):
        weight_test[i] = weight_z_phot(z_test[i],sigma_z_phot_test[i],z_cluster,sigma_z_cluster)
    weight_test_max = max(weight_test)
    weight_test_max_arg = argmax(weight_test)
    z_max = z_test[weight_test_max_arg]
    print 'max weight = {0}'.format(weight_test_max)
    print 'at redshift = {0}'.format(z_max)
    # calculate the weight of a photo-z galaxy at the cluster redshift
    sigma_z_phot_clsuter = s*(1+z_cluster)
    weight_z_phot_norm = weight_z_phot(z_cluster,sigma_z_phot_cluster,z_cluster,sigma_z_cluster)
    print 'weight_z_phot_norm = {0}'.format(weight_z_phot_norm)
    # determine the weights for all galaxies
    weight = numpy.zeros(N_cat)
    for i in range(N_cat):
        if cat[i,key['z_spec']] == -99:
            weight[i] = weight_z_phot(cat[i,key['z_mid']],sigma_z_phot[i],z_cluster,sigma_z_cluster)/weight_test_max*photo_z_penalty
        else:
            weight[i] = 1
    # set any weight values that might be negative = 0, these seem to be related to integration uncertainty
    mask_posweight = weight >= 0
    weight *= mask_posweight
    return weight
```
