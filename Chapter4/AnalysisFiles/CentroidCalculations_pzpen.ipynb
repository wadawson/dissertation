```python
# unpickle the catalogs
f = open(catalog1)
cat1 = pickle.load(f)
f.close()
f = open(catalog2)
cat2 = pickle.load(f)
f.close()
```

```python
# assign coordinate arrays
x1 = cat1[:,0]
y1 = cat1[:,1]
x2 = cat2[:,0]
y2 = cat2[:,1]
```

```python
# read in the fits files and their headers
hdulist1 = pyfits.open(fits1)
hdulist2 = pyfits.open(fits2)
# parse the WCS keywords in the primary HDU
wcs1 = wcs.WCS(hdulist1[0].header)
wcs2 = wcs.WCS(hdulist2[0].header)
# Convert pixel coordinates (x,y) to world coordinates (Ra, Dec)
# The second argument is "origin" -- in this case we're declaring we
# have 1-based coordinates.
skycord1 = wcs1.wcs_pix2sky(cat1, 1) 
skycord2 = wcs2.wcs_pix2sky(cat2, 1)
# assign world coordinate arrays
ra1 = skycord1[:,0]
dec1 = skycord1[:,1]
ra2 = skycord2[:,0]
dec2 = skycord2[:,1]
```

```python
# Transform the coordintes of each into a uniform arcsec difference system
ra_mean = numpy.mean(numpy.concatenate((numpy.reshape(ra1,(numpy.size(ra1),1)),numpy.reshape(ra2,(numpy.size(ra2),1)))))
dec_mean = numpy.mean(numpy.concatenate((numpy.reshape(dec1,(numpy.size(dec1),1)),numpy.reshape(dec2,(numpy.size(dec2),1)))))
```

```python
# calculate the angular separation from the mean
N_1 = numpy.size(ra1)
N_2 = numpy.size(ra2)
ra1_d = numpy.zeros(N_1)
dec1_d = numpy.zeros(N_1)
ra2_d = numpy.zeros(N_2)
dec2_d = numpy.zeros(N_2)
for i in numpy.arange(N_1):
    ra1_d[i], dec1_d[i] = tools.angcomp(ra_mean,dec_mean,ra1[i],dec1[i])
for i in numpy.arange(N_2):
    ra2_d[i], dec2_d[i] = tools.angcomp(ra_mean,dec_mean,ra2[i],dec2[i])
```

```python
# convert degrees to arcsec
ra1_d *= 60**2
dec1_d *= 60**2
ra2_d *= 60**2
dec2_d *= 60**2
```

```python
# Rotate the coordinates into a principle component system such that x captures
# the bulk of the centroid difference
ra1_d_mean = numpy.mean(ra1_d)
dec1_d_mean = numpy.mean(dec1_d)
ra2_d_mean = numpy.mean(ra2_d)
dec2_d_mean = numpy.mean(dec2_d)
theta = numpy.arctan((dec2_d_mean-dec1_d_mean)/(ra2_d_mean-ra1_d_mean))
x1_p = ra1_d*numpy.cos(theta)+dec1_d*numpy.sin(theta)
y1_p = dec1_d*numpy.cos(theta)-ra1_d*numpy.sin(theta)
x2_p = ra2_d*numpy.cos(theta)+dec2_d*numpy.sin(theta)
y2_p = dec2_d*numpy.cos(theta)-ra2_d*numpy.sin(theta)
```

```python
def histplot2d(x,y,x2,y2,prefix,prob=None,N_bins=100,histrange=None,x_lim=None,y_lim=None,x_label=None,y_label=None,legend=None):
    '''
    Input:
    x = [1D array of N floats]
    y = [1D array of N floats]
    prefix = [string] prefix of output file
    prob = [None] or [1D array of N floats] weights to apply to each (x,y) pair
    N_bins = [integer] the number of bins in the x and y directions   
    histrange = [None] or [array of floats: (x_min,x_max,y_min,y_max)] the range
        over which to perform the 2D histogram and estimate the confidence
        intervals
    x_lim = [None] or [array of floats: (x_min,x_max)] min and max of the range
        to plot
    y_lim = [None] or [array of floats: (x_min,x_max)] min and max of the range
        to plot
    x_label = [None] or [string] the plot's x-axis label
    y_label = [None] or [string] the plot's y-axis label
    legend = [None] or [True] whether to display a legend or not
    '''
    from matplotlib.colors import colorConverter
    #Create the confidence interval plot
    if histrange == None:
        if prob != None:
            H, xedges, yedges = numpy.histogram2d(x,y,bins=N_bins,weights=prob)
        elif prob == None:
            H, xedges, yedges = numpy.histogram2d(x,y,bins=N_bins)
    else:
        if prob != None:
            H, xedges, yedges = numpy.histogram2d(x,y,bins=N_bins,range=[[histrange[0],histrange[1]],[histrange[2],histrange[3]]],weights=prob)
        elif prob == None:
            H, xedges, yedges = numpy.histogram2d(x,y,bins=N_bins,range=[[histrange[0],histrange[1]],[histrange[2],histrange[3]]])
    H = numpy.transpose(H)
    #Flatten H
    h = numpy.reshape(H,(N_bins**2))
    #Sort h from smallest to largest
    index = numpy.argsort(h)
    h = h[index]
    h_sum = numpy.sum(h)
    # Find the 2 and 1 sigma levels of the MC hist
    for j in numpy.arange(numpy.size(h)):
        if j == 0:
            runsum = h[j]
        else:
            runsum += h[j]
        if runsum/h_sum <= 0.05:
            #then store the value of N at the 2sigma level
            h_2sigma = h[j]
        if runsum/h_sum <= 0.32:
            #then store the value of N at the 1sigma level
            h_1sigma = h[j]
    # Create the contour plot using the 2Dhist info
    # define pixel values to be at the center of the bins
    x = xedges[:-1]+(xedges[1]-xedges[0])/2
    y = yedges[:-1]+(yedges[1]-yedges[0])/2
    X, Y = numpy.meshgrid(x,y)

    #Create the confidence interval plot for dataset 2
    if histrange == None:
        if prob != None:
            H2, xedges2, yedges2 = numpy.histogram2d(x2,y2,bins=N_bins,weights=prob)
        elif prob == None:
            H2, xedges2, yedges2 = numpy.histogram2d(x2,y2,bins=N_bins)
    else:
        if prob != None:
            H2, xedges2, yedges2 = numpy.histogram2d(x2,y2,bins=N_bins,range=[[histrange[0],histrange[1]],[histrange[2],histrange[3]]],weights=prob)
        elif prob == None:
            H2, xedges2, yedges2 = numpy.histogram2d(x2,y2,bins=N_bins,range=[[histrange[0],histrange[1]],[histrange[2],histrange[3]]])
    H2 = numpy.transpose(H2)
    #Flatten H
    h2 = numpy.reshape(H2,(N_bins**2))
    #Sort h from smallest to largest
    index2 = numpy.argsort(h2)
    h2 = h2[index2]
    h_sum2 = numpy.sum(h2)
    # Find the 2 and 1 sigma levels of the MC hist
    for j in numpy.arange(numpy.size(h2)):
        if j == 0:
            runsum2 = h2[j]
        else:
            runsum2 += h2[j]
        if runsum2/h_sum2 <= 0.05:
            #then store the value of N at the 2sigma level
            h_2sigma2 = h2[j]
        if runsum2/h_sum2 <= 0.32:
            #then store the value of N at the 1sigma level
            h_1sigma2 = h2[j]
    # Create the contour plot using the 2Dhist info
    # define pixel values to be at the center of the bins
    x2 = xedges2[:-1]+(xedges2[1]-xedges2[0])/2
    y2 = yedges2[:-1]+(yedges2[1]-yedges2[0])/2
    X2, Y2 = numpy.meshgrid(x2,y2)

    
    fig = pylab.figure()
    # generate the colors for your colormap
    color11 = colorConverter.to_rgba('white')
    color21 = (8/255.,81/255.,156/255.)
#    color21 = colorConverter.to_rgba('black')
    color12 = colorConverter.to_rgba('white')
    color22 = (165/255.,15/255.,21/255.)
#    color22 = colorConverter.to_rgba('black')
    # make the colormaps
    cmap1 = mpl.colors.LinearSegmentedColormap.from_list('my_cmap',[color11,color21],256)
    cmap2 = mpl.colors.LinearSegmentedColormap.from_list('my_cmap2',[color12,color22],256)
    
    cmap1._init() # create the _lut array, with rgba values
    cmap2._init() # create the _lut array, with rgba values

    # create your alpha array and fill the colormap with them.
    # here it is progressive, but you can create whathever you want
    maxval = numpy.max((numpy.max(h),numpy.max(h2)))
    alphas1 = np.linspace(0, maxval, cmap1.N+3)
    cmap1._lut[:,-1] = alphas1
    alphas2 = np.linspace(0, maxval, cmap2.N+3)
    cmap2._lut[:,-1] = alphas2
    
    
#    pylab.plot(x,y,'b.',alpha=0.01,label='WL Centroid')
#    pylab.plot(x2,y2,'r.',alpha=0.01,label='Galaxy Centroid')

    # Countours
    CS = pylab.contour(X,Y,H,(h_2sigma,h_1sigma),linewidths=(2,2),colors=((158/255.,202/255.,225/255.),(49/255.,130/255.,189/255.)))
    CS2 = pylab.contour(X2,Y2,H2,(h_2sigma2,h_1sigma2),linewidths=(2,2),colors=((252/255.,146/255.,114/255.),(222/255.,45/255.,38/255.)))
    # imshow
    #im = pylab.imshow(H,cmap=pylab.cm.gray)
    pylab.pcolor(X,Y,H,cmap=cmap1)
    pylab.pcolor(X2,Y2,H2,cmap=cmap2)
    
    if x_label != None:
        pylab.xlabel(x_label,fontsize=14)
    if y_label != None:    
        pylab.ylabel(y_label,fontsize=14)
    if x_lim != None:
        pylab.xlim(x_lim)
    else:
        pylab.xlim((pylab.xlim()[1],pylab.xlim()[0]))
    if y_lim != None:
        pylab.ylim(y_lim)
    if legend != None:
        # Dummy lines for legend
        pylab.plot((0,1),(0,1),c='#800000',linewidth=2,label=('68%'))
        pylab.plot((0,1),(0,1),c='#0000A0',linewidth=2,label=('95%'))      
        pylab.legend(scatterpoints=1)
    fontsize=14
    ax = pylab.gca()
    for tick in ax.xaxis.get_major_ticks():
        tick.label1.set_fontsize(fontsize)
    for tick in ax.yaxis.get_major_ticks():
        tick.label1.set_fontsize(fontsize)
    
    filename = prefix+'_histplot2d'
    pylab.savefig(filename)
    
    return fig
```

```python
# Calculate the delx_p and dely_p arrays
# Now it will be nice if I reduce the separation in x' to see how the
# significance of the separation changes.
x1_p_mean = numpy.mean(x1_p)
y1_p_mean = numpy.mean(y1_p)
x2_p_mean = numpy.mean(x2_p)
y2_p_mean = numpy.mean(y2_p)
if x2_p_mean > x1_p_mean:
    delx_p = x2_p - x1_p
    dely_p = y1_p = y1_p
else:
    delx_p = x1_p - x2_p
    dely_p = y1_p - y1_p
delr_p = numpy.sqrt(delx_p**2+dely_p**2)
```

```python
# Now estimate the percent confidence as a function of delx_p and dely_p
# Order the del's in ascending order
idx_x = numpy.argsort(delx_p)
idx_y = numpy.argsort(dely_p)
idx_r = numpy.argsort(delr_p)
delx_p = delx_p[idx_x]
dely_p = dely_p[idx_y]
delr_p = delr_p[idx_r]
```

```python
# Create the cumulative step plot array
N_del = numpy.size(delx_p)
S = numpy.arange(N_del,0,-1)/N_del
```

```python
# Create the percent confidence plots
lw = 2
fig = pylab.figure()
pylab.step(delx_p,S)
#pylab.xlim((0,1))
pylab.ylim((0,1))
pylab.plot((0,0),(0,1),label='zero')
pylab.plot((6.3,6.3),pylab.ylim(),color='r',lw=lw,linestyle='-',label='$M_{gas}=5x10^{12}$')
pylab.plot((7.6,7.6),pylab.ylim(),color='r',lw=lw,linestyle='--',label='$M_{gas}=1x10^{13}$')
pylab.plot((10.2,10.2),pylab.ylim(),color='r',lw=lw,linestyle='-.',label='$M_{gas}=2.1x10^{13}$')
pylab.plot((3.4,3.4),pylab.ylim(),color='0.7',lw=lw,linestyle=':',label='$M_{gas}=0$')
pylab.legend(loc=0)
pylab.xlabel("$\Delta x'$ (arcsec)")
pylab.ylabel("Percent Confidence $>\Delta x'$")
#filename = prefix+'_delxPC'
#pylab.savefig(filename)
```
